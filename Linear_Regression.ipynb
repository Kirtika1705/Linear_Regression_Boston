```python
def linearity_test(data, target):
    corr_df = pd.DataFrame(columns= ['features', 'target', 'correlation', 'conclusion'])
    for cols in data.columns:
        r,p = pearsonr(data[cols], data[target])
        if p <0.05:
            corr_df = corr_df.append({'features':cols, 'target': target, 'correlation':r , 'conclusion':'significant'}, ignore_index=True)
        else:
            corr_df = corr_df.append({'features':cols, 'target': target, 'correlation':r , 'conclusion':'non-significant'}, ignore_index=True)
    return corr_df

linearity_test(data= df , target= 'medv')
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def VIF(df):
    VIF = pd.DataFrame()
    VIF["Feature"] = df.columns
    VIF["VIF"] = [variance_inflation_factor(df.values,i) for i in range(len(df.columns))]
    return VIF

VIF(df = x)
```
